```python
import numpy as np
import matplotlib.pyplot as plt
import astropy.io.fits as fits
```

```python
def display(fits_filename):
    """Display an image from a FITS file and save to a PDF file."""
    
    image = fits.getdata(fits_filename)
    plt.figure()
    plt.imshow(image, origin='lower', cmap='Greys_r')
    plt.colorbar()
    pdf_filename = fits_filename.replace('.fits', '.pdf')
    print('Writing: {0}'.format(pdf_filename))
    plt.savefig(pdf_filename)
```

## We see almost all black -- the image is not scaled properly! Now let's set the vmin and vmax image cut levels:

```python
# NOTE:  sigma_clipped_stats requires Astropy 1.0
from astropy.stats import sigma_clipped_stats
def display_scaled(fits_filename, cmap='Greys_r'):
    """Display an image from a FITS file and save to a PDF file."""
    
    image = fits.getdata(fits_filename)
    mean, median, stddev = sigma_clipped_stats(image, sigma=3.0)
    vmin = median - 3.*stddev
    vmax = median + 3.*stddev
    plt.figure()
    plt.imshow(image, origin='lower', cmap=cmap, vmin=vmin, vmax=vmax)
    plt.colorbar()
    pdf_filename = fits_filename.replace('.fits', '.pdf')
    print('Writing: {0}'.format(pdf_filename))
    plt.savefig(pdf_filename)
```

```python
display_scaled(fits_filename, cmap='Blues_r')
```

## NEW: Astropy 1.0 introduced the `visualization` module

```python
# one way - scale the image directly
from astropy.visualization import scale_image
image = fits.getdata(fits_filename)
plt.imshow(scale_image(image, scale='sqrt', percent=99.), origin='lower', cmap='hot')
plt.colorbar()
```

```python
# another way - input norm into imshow()
# (this gives colorbars with real image values)
from astropy.visualization import PercentileInterval, SqrtStretch
from astropy.visualization.mpl_normalize import ImageNormalize
vmin, vmax = PercentileInterval(99.).get_limits(image)
norm = ImageNormalize(vmin=vmin, vmax=vmax, stretch=SqrtStretch(), clip=True)
plt.imshow(image, origin='lower', cmap='hot', norm=norm)
plt.colorbar(ticks=np.arange(0, 61, 10))
```

```python
%%file mymodule.py
import os
from astropy.stats import sigma_clipped_stats
import astropy.io.fits as fits
import matplotlib.pyplot as plt
from pyraf import iraf
from iraf import noao, digiphot, daophot, daofind


__author__ = 'Larry Bradley'


def display_scaled(fits_filename, cmap='Greys_r'):
    """
    Display an image from a FITS file and save to a PDF file.
    
    Example
    -------
    >>> display_scaled('myfile.fits', cmap='hot')
    """
    
    image = fits.getdata(fits_filename)
    mean, median, stddev = sigma_clipped_stats(image, sigma=3.0)
    vmin = median - 3.*stddev
    vmax = median + 3.*stddev
    plt.figure()
    plt.imshow(image, origin='lower', cmap=cmap, vmin=vmin, vmax=vmax)
    plt.colorbar()
    pdf_filename = fits_filename.replace('.fits', '.pdf')
    print('Writing: {0}'.format(pdf_filename))
    plt.savefig(pdf_filename)

    
def find_objects(fits_filename):
    """
    Find stars in an image using DAOFIND.
    
    Example
    -------
    coord_fn = find_objects('myfile.fits')
    """

    out_filename = fits_filename + '.stars'
        
    # remove out_filename if it already exists
    if os.path.isfile(out_filename):
        print('Removing previous star coordinates file {0}'.format(out_filename))
        os.remove(out_filename)
        
    # define DAOFIND parameters
    iraf.unlearn('daofind')
    iraf.unlearn('datapars')
    iraf.unlearn('findpars')
    iraf.datapars.sigma = 1.0       # you should use the real background sigma
    iraf.datapars.fwhmpsf = 2.5
    iraf.daofind.boundary = 'constant'
    iraf.daofind.constant = 0.
    iraf.daofind.findpars.threshold = 3.0    # 3-sigma detections only
    iraf.daofind.findpars.nsigma = 1.5    # width of convolution kernel in sigma
    iraf.daofind.findpars.ratio = 1.0    # ratio of Gaussian axes
    iraf.daofind.findpars.theta = 0.
    iraf.daofind.findpars.sharplo = 0.2    # lower bound on sharpness
    iraf.daofind.findpars.sharphi = 1.0    # upper bound on sharpness
    iraf.daofind.findpars.roundlo = -1.0    # lower bound on roundness
    iraf.daofind.findpars.roundhi = 1.0    # upper bound on roundness
    iraf.daofind.findpars.mkdetections = "no"
    
    # run DAOFIND
    in_filename = fits_filename + '[SCI, 1]'  # use the first science extension for this example
    iraf.daofind(image=in_filename, output=out_filename, interactive='no', verify='no', verbose='no')

    print('Saved star coordinates to {0}'.format(out_filename))
    
    return out_filename    # return the name of the saved file
```
