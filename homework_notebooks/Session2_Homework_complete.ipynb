## Problem #1

1.  Create three data arrays.  The first should consist of values drawn from a normal distribution with a mean of 100.0 and a standard deviation of 10.0.  The second should consist of numbers that are equal to the square root of the values in the first array if positive, 0 of negative.  The third should be an array of integers that are all 0 except where the values in the first array are more than 2 standard deviations from the mean.  If they are > 2sd from the mean, array3 = 2, if < 2sd from the mean, array3 = 4.

  All arrays should be 1024 x 1024 in size.

  Array 1 should be Int16, 2 should be Float32 and 3 should be Int16

  Verify that the first array does indeed have a mean of 100 and a standard deviation of 10.

  Write out all 3 arrays to  FITS file with array 1 in extension 1, array 2 in extension 2 and array 3 in extension 3.  Include a keyword in the primary header MYNAME and set it equal to a string with your name.

```python
import numpy as np
from astropy.io import fits
```

The first thing I did was google "numpy normal", and it suggested "numpy normal distribution" for me.  Clicking on this link led me to this page:

http://docs.scipy.org/doc/numpy/reference/generated/numpy.random.normal.html


```python
array1 = np.random.normal(loc=100.0, scale=10.0, size=(1024,1024))
```

```python
array1
```

```python
array1.mean()
```

```python
array1.std()
```

This is floating point of course, and I want short integer type.  I don't want to just do integer truncation, because that'll bias the mean.  What I want is the nearest integer.  Google was my friend again, pointing me to this page

http://docs.scipy.org/doc/numpy/reference/routines.math.html

that mentions the nearest integer function.  I use the 'astype' method of numpy arrays to cast the array to int16.

```python
a1 = np.rint(array1).astype(np.int16)
```

Notice how you can chain methods - np.rint(array1) returns a numpy ndarray, so I can use the astype() method on that object to get my final array.

```python
a1
```

```python
a1.mean()
```

```python
a1.std()
```

Check to see if there's any negative values (there shouldn't be...)

```python
np.where(a1<0)
```

Since there aren't any negative values, it's safe to just take the square root.

```python
a2 = np.sqrt(a1)
```

```python
a2.mean()
```

This isn't equal to 10.0 - why?

```python
a2
```

For the third array, all the elements with a value between (mean - 2\*sigma) and (mean + 2\*sigma) will be zero, so it's a good idea to start with an array of zeros.  Numpy provides this (as well as np.ones() if you want an array of 1s)

```python
a3 = np.zeros((1024,1024)).astype(np.int16)
```

```python
a3
```

```python
fours = np.where(a1 <= (a1.mean() - 2.0*a1.std()))
twos = np.where(a1 >= (a1.mean() + 2.0*a1.std()))
```

```python
a3[fours] = 4
a3[twos] = 2
```

```python
np.where(a3==2)[0].size
```

```python
np.where(a3==4)[0].size
```

These should both be about 2.3% of the total number of pixels.


Now to make the FITS file.  The FITS documentation at

http://astropy.readthedocs.org/en/latest/io/fits/ 

gives some info about how to do this, but unless you want to do exactly what their example shows (making a FITS file with a single PrimaryHDU with data), it can be confusing.  I'll try and fix that.  Meanwhile, here's what you need to know.

There's really two ways to go.  You can start by making an empty HDUList, then  create each HDU separately, and append them into your HDUList.  Alternatively, you can create each of your HDUs, then create the HDUList from the HDUs.  I'll go through both ways.

In the first way, we start by making an empty HDUList

```python
f1 = fits.HDUList()
```

Notice we just called the HDUList class name with no arguments in the parentheses.  When you make an instance of a class (f1 is our instance of the class 'fits.HDUList'), you are calling the class's *constructor*.  The HDUList class has a function (if you really want to know, it's name is \_\_init\_\_) that gets called when you create an instance of a class.  This function handles the initialization of the members of the class from whatever arguments are included (if necessary), and can also handle making sure that the arguments are of the appropriate type.

```python
f1
```

f1 appears to be an empty list when we output it to the screen, but it's really not:

```python
type(f1)
```

Now we add a Primary HDU to our HDUList.  Again, we don't have anything to put in our PrimaryHDU, so we just make an empty one by not specifying anything in the parentheses.  We use the 'append' method of the HDUList object (see http://astropy.readthedocs.org/en/latest/io/fits/api/hdulists.html#hdulist
for more details about the HDUList object).

```python
f1.append(fits.PrimaryHDU())
```

Create three ImageHDUs

```python
imh1 = fits.ImageHDU()
imh2 = fits.ImageHDU()
imh3 = fits.ImageHDU()
```

The 3 ImageHDUs we created are distinct objects.  We can attach the arrays we want to the .data members of each:

```python
imh1.data = a1
imh2.data = a2
imh3.data = a3
```

And now we can append each of these ImageHDUs to our HDUList

```python
f1.append(imh1)
f1.append(imh2)
f1.append(imh3)
```

Alternatively, we could have created our PrimaryHDU and 3 ImageHDUs first, then created the HDUList from them:

```python
myhdulist = fits.HDUList(hdus=[fits.PrimaryHDU(),imh1, imh2, imh3])
```

```python
f1[0].header
```

Add the required keyword/value pair

```python
f1[0].header['MYNAME'] = 'Robert Jedrzejewski'
```

And write out to a FITS file

```python
f1.writeto('Session2_Q1.fits',clobber=True)
```

Check that everything seems OK

```python
fits.info('Session2_Q1.fits')
```

## Problem #3

Make a table with 8 columns:

    Theta (degrees)
    Theta (radians)
    Sin(Theta)
    Cos(Theta)
    Tan(Theta)
    Csc(Theta)
    Sec(Theta)
    Cot(Theta)

Populate this table with the correct values for Theta in degrees from 0 to 360.0 in increments of 1.0

If the data overflows, replace with 99999.0 or -99999.0

Write out the table in Latex form.

```python
from astropy.io import ascii
```

First make an array for the x values

```python
degrees = np.arange(361).astype(np.float32)
```

```python
degrees
```

```python
import math
radians = degrees * math.pi / 180.0
```

Now do the math

```python
sintheta = sin(radians)
costheta = cos(radians)
tantheta = tan(radians)
sectheta = 1.0/cos(radians)
csctheta = 1.0/sin(radians)
cottheta = 1.0/tan(radians)
```

We got a couple of divide by zero warnings.  They are for csc and cot evaluated at 0.  The other asymptotic values don't give an error because they would rely on the exact value of pi.

We can fix them by just setting values >99999.0 to 99999.0, and values less than -99999.0 to -99999.0.

```python
tantheta[np.where(tantheta > 99999.0)] = 99999.0
tantheta[np.where(tantheta < -99999.0)] = -99999.0
sectheta[np.where(sectheta > 99999.0)] = 99999.0
sectheta[np.where(sectheta < -99999.0)] = -99999.0
csctheta[np.where(csctheta > 99999.0)] = 99999.0
csctheta[np.where(csctheta < -99999.0)] = -99999.0
cottheta[np.where(cottheta > 99999.0)] = 99999.0
cottheta[np.where(cottheta < -99999.0)] = -99999.0
```

Now write out the table.  The astropy.ascii documentation tells us that to write out in Latex format, we just need to choose Writer=ascii.Latex in the ascii.write call

```python
columns = [degrees, radians, sintheta, costheta, tantheta, csctheta, sectheta, cottheta]
names = ['Angle (degrees)', 'Angle (radians)', 'sin()', 'cos()', 'tan()', 'csc()', 'sec()', 'cot()']
ascii.write(columns, names=names, output='Session2_Q3.tex', Writer=ascii.Latex)
```

Notice that I didn't use the astropy.table package at all, the astropy.io.ascii package made it easy to just create the columns and names, then write it out to a Latex-style text file
